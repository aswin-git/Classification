## Importing Essential Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Loading the Breast Cancer Dataset

```python
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()
x = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

x stores the features in a pandas DataFrame with meaningful column names.

y stores the target variable — a binary classification label (0 = malignant, 1 = benign)

```python
yl = pd.Series(data.target, name='target')
df = x.join(yl)
```

##  Train-Test Split

X_train, y_train: Used to train the model (80% of the data).

X_test, y_test: Used to test the model’s performance (20% of the data).



```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Feature Scaling with StandardScaler


StandardScaler standardizes features by removing the mean and scaling to unit variance.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit only on training data, then transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# Feature Selection using Lasso Regularization

LassoCV: Performs L1 regularization with cross-validation, shrinking less important feature coefficients to zero.

selected_features: Indices of features that were retained (non-zero coefficients).

This results in a dimensionality reduction, enhancing both efficiency and model performance.

```python
from sklearn.linear_model import LassoCV


lasso = LassoCV(cv=5, random_state=0)
lasso.fit(X_train_scaled, y_train)

# Identify selected features (non-zero coefficients)
selected_features = np.where(lasso.coef_ != 0)[0]
X_train_selected = X_train_scaled[:, selected_features]
X_test_selected = X_test_scaled[:, selected_features]

print(f"Selected Features Index: {selected_features}")
print(f"Number of Features Reduced From {x.shape[1]} to {X_train_selected.shape[1]}")
```

# Logistic Regression Model Training and Evaluation


```python
from sklearn.linear_model import  LogisticRegression
from sklearn.metrics import classification_report, accuracy_score



# Logistic Regression
clf = LogisticRegression(max_iter=500, random_state=42)
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

#  Pipeline with Model Comparison and Hyperparameter Tuning

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

models = {
    "Logistic Regression": {
        "pipeline": Pipeline([
            ('scaler', StandardScaler()),
            ('clf', LogisticRegression(max_iter=1000))
        ]),
        "params": {
            "clf__C": [0.1, 1, 10]
        }
    },
    "Decision Tree": {
        "pipeline": Pipeline([
            ('scaler', StandardScaler()),  # Optional for trees, but can keep for uniformity
            ('clf', DecisionTreeClassifier())
        ]),
        "params": {
            "clf__criterion": ["gini", "entropy"],
            "clf__max_depth": [3, 5, 10]
        }
    },
    "Random Forest": {
        "pipeline": Pipeline([
            ('scaler', StandardScaler()),  # Optional but harmless here
            ('clf', RandomForestClassifier())
        ]),
        "params": {
            "clf__n_estimators": [5, 10, 20, 35, 50, 100],
            "clf__max_depth": [5, 10]
        }
    },
    "SVM": {
        "pipeline": Pipeline([
            ('scaler', StandardScaler()),
            ('clf', SVC())
        ]),
        "params": {
            "clf__C": [0.1, 1, 10],
            "clf__kernel": ["linear", "rbf"]
        }
    },
    "KNN": {
        "pipeline": Pipeline([
            ('scaler', StandardScaler()),
            ('clf', KNeighborsClassifier())
        ]),
        "params": {
            "clf__n_neighbors": [3, 5, 7]
        }
    }
}

results = []

for name, mp in models.items():
    print(f"Running GridSearchCV for {name}...")
    
    clf = GridSearchCV(mp['pipeline'], mp['params'], cv=5, scoring='accuracy', n_jobs=-1)
    clf.fit(X_train_selected, y_train)
    y_pred = clf.predict(X_test_selected)
    
    results.append({
        "Model": name,
        "Best Params": clf.best_params_,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred, zero_division=0),
        "Recall": recall_score(y_test, y_pred, zero_division=0),
        "F1 Score": f1_score(y_test, y_pred, zero_division=0)
    })

results_df = pd.DataFrame(results)
results_df = results_df.sort_values(by="F1 Score", ascending=False)

print(results_df)


```

### Insights
SVM (RBF kernel) emerged as the top performer with an F1 Score of 0.985, showcasing an excellent balance of precision and recall.

Logistic Regression delivered stable performance with strong metrics across the board and less complexity compared to tree-based models.

KNN achieved perfect recall but slightly lower precision, which might be useful in recall-sensitive applications (e.g., medical diagnosis).

Random Forest offered balanced performance but slightly lagged behind in F1 Score compared to SVM and KNN.

Decision Tree, while interpretable, had the lowest overall performance, particularly in recall, indicating possible overfitting or underfitting at the chosen depth.
